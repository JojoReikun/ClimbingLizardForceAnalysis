```python
#### GET MEANS of angle to y-vector in degree
#### CW rot = pos values; CCW rot = neg values
variable = "MeanXY"
print("UP: \n")

Fxy_up_fr = data.loc[(data["direction"] == "up") & (data["foot"] == "FR")]
print("mean_up_fr: ", np.nanmean(Fxy_up_fr[variable]))
print("sd_up_fr: ", np.nanstd(Fxy_up_fr[variable]))
print("median_up_fr: ", np.nanmedian(Fxy_up_fr[variable]))

Fxy_up_fl = data.loc[(data["direction"] == "up") & (data["foot"] == "FL")]
print("\nmean_up_fl: ", np.nanmean(Fxy_up_fl[variable]))
print("sd_up_fl: ", np.nanstd(Fxy_up_fl[variable]))
print("median_up_fl: ", np.nanmedian(Fxy_up_fl[variable]))

Fxy_up_hr = data.loc[(data["direction"] == "up") & (data["foot"] == "HR")]
print("\nmean_up_hr: ", np.nanmean(Fxy_up_hr[variable]))
print("sd_up_hr: ", np.nanstd(Fxy_up_hr[variable]))
print("median_up_hr: ", np.nanmedian(Fxy_up_hr[variable]))

Fxy_up_hl = data.loc[(data["direction"] == "up") & (data["foot"] == "HL")]
print("\nmean_up_hl: ", np.nanmean(Fxy_up_hl[variable]))
print("sd_up_hl: ", np.nanstd(Fxy_up_hl[variable]))
print("median_up_hl: ", np.nanmedian(Fxy_up_hl[variable]))

print("\nDOWN: \n")

Fxy_down_fr = data.loc[(data["direction"] == "down") & (data["foot"] == "FR")]
print("mean_down_fr: ", np.nanmean(Fxy_down_fr[variable]))
print("sd_down_fr: ", np.nanstd(Fxy_down_fr[variable]))
print("median_down_fr: ", np.nanmedian(Fxy_down_fr[variable]))

Fxy_down_fl = data.loc[(data["direction"] == "down") & (data["foot"] == "FL")]
print("\nmean_down_fl: ", np.nanmean(Fxy_down_fl[variable]))
print("sd_down_fl: ", np.nanstd(Fxy_down_fl[variable]))
print("median_down_fl: ", np.nanmedian(Fxy_down_fl[variable]))

Fxy_down_hr = data.loc[(data["direction"] == "down") & (data["foot"] == "HR")]
print("\nmean_down_hr: ", np.nanmean(Fxy_down_hr[variable]))
print("sd_down_hr: ", np.nanstd(Fxy_down_hr[variable]))
print("median_down_hr: ", np.nanmedian(Fxy_down_hr[variable]))

Fxy_down_hl = data.loc[(data["direction"] == "down") & (data["foot"] == "HL")]
print("\nmean_down_hl: ", np.nanmean(Fxy_down_hl[variable]))
print("sd_down_hl: ", np.nanstd(Fxy_down_hl[variable]))
print("median_down_hl: ", np.nanmedian(Fxy_down_hl[variable]))
```
